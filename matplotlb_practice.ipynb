```python
%matplotlib inline
```

```python
#Import the Dependencies
import matplotlib.pyplot as plt
```

```python
#Sets the X-axis to a list of strings for each month
x_axis = ["Jan","Feb","Mar","Apr","May","Jun","Jul","Aug","Sep","Oct","Nov","Dec"]

#Sets the Y-axis to a list of floats as the total fare of U.S dollars accumulated per month
y_axis = [10.02,23.24,39.20,35.42,32.34,27.04,43.82,10.56,11.85,27.90,20.71,20.09]
```

```python
#Construct the plot
plt.plot(x_axis,y_axis)
plt.show()
```

```python
# Create the plot with ax.plt()
fig, ax =plt.subplots()
ax.plot(x_axis, y_axis)
```

```python
#Create the plot with ax.plt()
fig = plt.figure()
ax = fig.add_subplot()
ax.plot(x_axis,y_axis)
```

```python
#Create the Plot and add a label for the legend
plt.plot(x_axis,y_axis, label = 'Boston')
# Create labels for the x and y axes.
plt.xlabel("Date")
plt.ylabel("Fare($)")
#Set y limit b/t 0-45
plt.ylim(0,45)
#Create a title
plt.title("PyBer Fare by Month")
#Add the legend
plt.legend()
plt.show()
```

```python
fig, ax = plt.subplots()
ax.plot(x_axis,y_axis,marker = "d",linewidth=2,label = 'Boston',color = "green")
ax.set_xlabel('Date')
ax.set_ylabel('Fare($)')
ax.grid()
ax.set_title("PyBer Fare by Month")
ax.legend()
ax.set_ylim(0,45)
plt.show()
```

```python
#Create bar
plt.bar(x_axis,y_axis)
```

```python
#create the plot.
plt.bar(x_axis, y_axis, color = "green", label = 'Boston')
#Create labels for x and y axes.
plt.xlabel("Date")
plt.ylabel("Fare($)")
#Create title
plt.title("PyBer Fare by Month")
#Add legend
plt.legend()
plt.show()
```

```python
#Create the plot
plt.barh(x_axis, y_axis,color = "magenta",label = "Boston")
plt.ylabel("Date")
plt.xlabel("Fare($)")
plt.title("PyBer Fare by Month")
plt.gca().invert_yaxis()
plt.show()
```

```python
#Create plot with ax.plt()
fir, ax = plt.subplots()
ax.bar(x_axis, y_axis)
plt.show()
```

```python
#Create plot with ax.plt()
fig, ax = plt.subplots()
ax.barh(x_axis, y_axis,color="cyan",label = "Chicago")
ax.legend()
ax.set_xlabel("Fare($)")
ax.set_ylabel("Date")
ax.set_title("PyBer Fare by Month")
plt.show()
```

```python
plt.plot(y_axis,x_axis,'o',color = "red",label = "Chicago")
plt.title("PyBer Fare by Month")
plt.xlabel("Fare($)")
plt.ylabel("Date")
plt.gca().invert_yaxis()
plt.show()
```

```python
plt.scatter(x_axis, y_axis)
plt.show()
```

```python

y_axis_larger = []
for data in y_axis:
    y_axis_larger.append(data*3)
plt.scatter(x_axis, y_axis, s=y_axis_larger)
plt.show()
```

```python
fig, ax = plt.subplots()
ax.scatter(x_axis, y_axis)
plt.show()
```

```python
fig, ax = plt.subplots()
y_axis_larger = []
for data in y_axis:
    y_axis_larger.append(data*5)
ax.scatter(y_axis,x_axis,s=y_axis_larger,color = "skyblue",alpha = .2,edgecolors="black",linewidth = 2,label = 'Boston')
ax.legend()
ax.set_xlim(0,50)
ax.set_title("PyBer Fare by Month")
ax.set_xlabel("Fare($)")
ax.set_ylabel("Date")
ax.invert_yaxis()
plt.show()
```

```python
plt.pie(y_axis, labels = x_axis)
plt.show()
```

```python
plt.subplots(figsize=(8, 8))
explode_values = (0, 0, 0, 0, 0, 0, 0.2, 0, 0, 0, 0, 0)
plt.pie(y_axis, explode = explode_values, labels = x_axis, autopct = '%.1f%%')
plt.show()
```

```python
#Assign 12 colors, one for each month
colors = ["slateblue", "magenta","lightblue","green","yellowgreen","greenyellow", "yellow", "orange", "gold", "indianred", "tomato", "mistyrose"]
explode_values = (0, 0, 0, 0, 0, 0, 0.2, 0, 0, 0, 0, 0)
plt.subplots(figsize=(8,8))
plt.pie(y_axis, explode = explode_values,colors = colors, labels = x_axis, autopct = '%.1f%%')
plt.show()
```

```python
fig, ax = plt.subplots(figsize=(8,8))
explode_values = (0, 0, 0.2, 0, 0, 0, 0.2, 0, 0, 0, 0, 0)
colors = ["slateblue", "magenta","lightblue","green","yellowgreen","greenyellow", "yellow", "orange", "gold", "indianred", "tomato", "mistyrose"]

ax.pie(y_axis,labels = x_axis,autopct = "%.1f%%",explode = explode_values,shadow = True,
      startangle=95,counterclock=False,colors=colors)
plt.show()
```
