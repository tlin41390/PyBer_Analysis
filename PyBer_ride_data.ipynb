```python
%matplotlib inline
#Dependencies
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import statistics
#Load in the csv files
pyber_ride_df = pd.read_csv("Resources/PyBer_ride_data.csv")
pyber_ride_df
```

```python

#Set x-axis and tick locations
x_axis = np.arange(len(pyber_ride_df))
tick_locations = [value for value in x_axis]
#Plot data
pyber_ride_df.plot(x="Month",y="Avg. Fare ($USD)")
plt.xticks(tick_locations,pyber_ride_df["Month"])
plt.show()
```

```python
pyber_ride_df.plot.bar(x="Month",y="Avg. Fare ($USD)")
plt.show()
```

```python
pyber_ride_df.plot(x="Month",y="Avg. Fare ($USD)", kind = 'bar')
plt.show()
```

```python
y_axis = pyber_ride_df.loc[:,'Avg. Fare ($USD)']
stdev = statistics.stdev(y_axis)
pyber_ride_df.describe()
pyber_ride_df.plot(x ="Month",y="Avg. Fare ($USD)",kind = 'bar',color = "skyblue",capsize=3,yerr = stdev)
plt.yticks(np.arange(0,51,step = 5.0))
plt.xticks(rotation = 0)
plt.show()
```
